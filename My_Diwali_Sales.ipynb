# Data Cleaning

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
# "any"  paratmeter is used to removed entire row if any single value is NA
df.dropna(how='any',inplace=True,)
df
```

```python
df.isnull().sum()
```

```python

df['Amount'] = df['Amount'].astype('int')
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status' : 'Shaadi'})
```
